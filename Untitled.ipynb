```python
def remove_linear_background(stack, left_width=20, right_width=20):
    """
    Remove linear background along columns for each image in a stack.
    
    Parameters
    ----------
    stack : ndarray (N, H, W)
        Image stack.
    left_width : int
        Width of left background region.
    right_width : int
        Width of right background region.
    
    Returns
    -------
    corrected : ndarray (N, H, W)
        Background-corrected image stack.
    """

    N, H, W = stack.shape
    corrected = np.empty_like(stack, dtype=np.float32)

    # Precompute x positions for linear interpolation
    x = np.linspace(0, 1, W, dtype=np.float32)

    for i in range(N):
        img = stack[i]

        # Compute row-wise left and right means
        left_mean  = img[:, :left_width].mean(axis=1)      # shape (H,)
        right_mean = img[:, -right_width:].mean(axis=1)    # shape (H,)

        # Compute slope for each row
        slope = (right_mean - left_mean)[:, None]          # shape (H,1)

        # Background model for each row
        background = left_mean[:, None] + slope * x[None, :]

        # Subtract background
        corrected[i] = img - background

    return corrected
```
